# Linear Regression 

Linear Regression is a machine learning algorithm based on **supervised learning**. It performs a regression task. 

# Importing Libraries

```python
import pandas as pd
import numpy as np
```

# Reading Dataset

```python
df = pd.read_csv('../input/ice-cream-revenue/IceCreamData.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
X = df['Temperature']
Y = df['Revenue']
```

# Now Spliting data in test and train data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y , test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

Object Creation

```python
model = LinearRegression()
```

Training of model

```python
len(X_train)
```

```python
len(Y_train)
```

```python
len(X_test)
```

```python
len(Y_test)
```

```python
X_train.ndim
```

```python
X_train = np.array([X_train])
```

```python
X_train.ndim
```

```python
model.fit(X_train.reshape(-1,1), Y_train)
```

```python
X_test.ndim
```

```python
X_test = np.array([X_test])
```

Now model ios created and we have to pass the test data for prediction

```python
Predict = model.predict(X_test.reshape(-1 ,1))
```

```python
Predict
```

```python
Y_test
```

# Check Accuracy

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_test, Predict)
```
